```python
import numpy as np
import pandas as pd

x_train = pd.read_csv("x_filter.csv", index_col = 0)
y_train = pd.read_csv("y_train_use2.csv", index_col = 0)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.decomposition import PCA
import matplotlib as plt
import matplotlib.pyplot as plt_py
```

```python
#kmeans_3 model and centroids
kmeans3 = KMeans( n_clusters=3, init='k-means++' ) 
kmeans3.fit( x_train, y_train )
```

```python
x_train[ 'kmeansCluster_3' ] = kmeans3.labels_
centroids3 = kmeans3.cluster_centers_
```

```python
#kmeans_5 model and centroids
kmeans5 = KMeans( n_clusters=5, init='k-means++' ) 
kmeans5.fit( x_train, y_train )
x_train[ 'kmeansCluster_5' ] = kmeans5.labels_
centroids5 = kmeans5.cluster_centers_
```

```python
#kmeans_8 model and centroids
kmeans8 = KMeans( n_clusters=8, init='k-means++' ) 
kmeans8.fit( x_train, y_train )
x_train[ 'kmeansCluster_8' ] = kmeans8.labels_
centroids8 = kmeans8.cluster_centers_
```

```python
#kmeans_11 model and centroids
kmeans11 = KMeans( n_clusters=11, init='k-means++' ) 
kmeans11.fit( x_train, y_train )
x_train[ 'kmeansCluster_11' ] = kmeans11.labels_
centroids11 = kmeans11.cluster_centers_
```
