```python
import numpy as np
import pandas as pd
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

```python
x_train = pd.read_csv('x_train_oct.csv')
y_train = pd.read_csv('y_train_oct.csv')
df = pd.concat([x_train, y_train], axis=1)
```

```python
class_counts=df.iloc[:,-1].value_counts()
```

```python
min_class = class_counts.idxmin()
min_count = class_counts.min()
```

```python
balanced_data = pd.DataFrame()
for class_index in class_counts.index:
    class_subset = df[df.iloc[:,-1] == class_index]
    resampled_subset = resample(class_subset,
                                replace=False,
                                n_samples=min_count,
                                random_state=123)  # Ensure reproducibility
    balanced_data = pd.concat([balanced_data, resampled_subset], axis=0)
```

```python
topRows=[293,266,349,322,238,-1]
newData = balanced_data.iloc[:,topRows]
```

```python
datameans = newData.groupby(df.iloc[:, -1]).mean()
datameans
```

```python
N = 5
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # close the loop
```

```python
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], ['293','266','349','322','238'])

#class 0
values = datameans.iloc[0,:-1].tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Class 0')

#class 1
values = datameans.iloc[1,:-1].tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Class 1')

#class 2
values = datameans.iloc[2,:-1].tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Class 2')

#class 3
values = datameans.iloc[3,:-1].tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Class 3')

plt.legend()
```
